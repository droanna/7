1. Obliczenie orientacyjnej ceny mieszkania za 5 lat

```python
import numpy as np

init_price = 120000
period = 5
increase_rate = 0.05

prices = init_price * (1 + increase_rate) ** np.arange(period + 1)
price = round(prices[5], 2)
print(price)
```

2. Obliczenie wysokości miesięcznej wpłaty na lokatę

```python
import numpy_financial as npf

rate = 0.12/12
nper = period*12


payment = npf.pmt(rate, nper, pv=0, fv=price, when='end')
payment = round(-payment, 2)
payment
```

3. Wykres zmian ceny mieszkania i wartości odłożonej sumy

```python
prices = np.arange(nper) * (price - init_price)/(nper - 1) + init_price
payments = payment * (1 + rate) ** np.arange(nper + 1)
```

```python
import matplotlib.pyplot as plt

plt.plot(prices,label='cena mieszkania')
plt.plot(payments.cumsum(),label='wpłaty')
plt.legend()
plt.xlabel('Liczba okresów')
plt.ylabel('Wartość')
```
